```python
# Assignment 3
# Creating empty set
set1 = set()
# Generating the first 5 numbers to be inputted in the set
for i in range(5):
  set1.add(i)
print(set1)

# Creating empty set 2
set2 = set()
# Generating even numbers to be in the set 2
for j in range(2,11,2):
  set2.add(j)
print(set2)

# Union of the sets
union_set = set1.union(set2)
print(union_set)
# Intersection of Sets
intersection_set = set1.intersection(set2)
print(intersection_set)
# Difference of set
# Elements in set 2 not in set1
difference_set = set2.difference(set1)
print(difference_set)
# Symmetrical intersection of sets
symmertric_difference = set1.symmetric_difference(set2)
print(symmertric_difference)
```
