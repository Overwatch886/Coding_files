```python
# Assignment 14
original_lst = [1,3,"pay", 'drag', True, False]
new_lst = []
n = int(input("Enter Number of Positions:"))
no_of_rotations = int(len(original_lst)/n)
for i in range(no_of_rotations):
    new = original_lst[-n:]
    new_lst.append(new)
    original_lst = original_lst[-len(original_lst):-n]
print(new_lst)
```
