```python
# Assignment 1
# Creating a dictionary with first 10 integers as keys and their squares as its values
dictionary = dict()
for x in range(1,11):
    dictionary[x] = x**2
print(dictionary)
```

```python
# Assignment 2
# Accessing dictionary elements
print(dictionary[5])
print(dictionary.keys())
```

```python
# Assignment 3
# Dictionay methods
dictionary[11] = 121
print(dictionary)
dictionary.pop(1)
print(dictionary)
```

```python
# Assignment 4
# Iterating Over Dictionaries
for key,value in dictionary.items():
    print(f"{key} : {value}")
```

```python
# Assignment 5 
# Dictionary comprehension
dictionary = {x : x**3 for x in range(1,11)}
print(dictionary)
```

```python
# Assignment 6
# Merging dictionaries
dict1 = {x : x**2 for x in range(1, 6)}
dict2 = {x : x**2 for x in range(5, 11)}
merged_dict = dict1.copy()
merged_dict.update(dict2)
print(merged_dict)
```

```python
# Assignment 6
# Merging dictionaries
dict1 = {x : x**2 for x in range(1, 6)}
dict2 = {x : x**2 for x in range(5, 11)}
merged_dict = {**dict1, **dict2}
print(merged_dict)
```

```python
# Assignment 7 
# Nested Dictionaries
nested_dictionary = {'name' : 'Israel', 'age' : 16, 'grade' : {'math' : 'A1', 'science' : 'B2', 'english' : 'C5'}}
print(nested_dictionary)
```

```python
# Assignment 8
# Dictionary of Lists
dictionary = {x: [x * i for i in range(1, 6)] for x in range(1, 6)}
print(dictionary)
```

```python
# Assignment 9
dictionary = {x : (x, x**2) for x in range(1,6)}
print(dictionary)
```

```python
# Assignment 10
# Dictionary and List Conversion
dictionary = {x : x**2 for x in range (1,6)}
list_of_tuples = [(key,value) for (key, value) in dictionary.items()]
print(list_of_tuples)
```

```python
# Assignment 10
list_of_tuples = list(dictionary.items())
print(list_of_tuples)
```

```python
# Assignment 11
# Dictionary Filtering
dictionary = {x : x**2 for x in range (1,11)}
new_dictionary = {x : x**2 for x in range (1,11) if x%2 == 0}
print(new_dictionary)
```

```python
# Assignment 12
# Dictionary key and value Transformation
dictionary = {x : x**2 for x in range (1,6)}
another_dictionary = {v : k for k,v in dictionary.items()}
print(another_dictionary)
```

```python
# Assignment 13 
# Default Dictionary
keys = ['a','b' ,'c' ,'d']
values = {value : [] for value in keys}
values['a'].append('alpha')
values['b'].append('beta')
values['c'].append('caeser')
values['d'].append('dollar')
print(values)
```

```python
# Assignment 14
# Counting with dictionaries
def counting_dicts(a_string): 
    letters = list(a_string)
    dictionary = dict()
    for letter in letters:
        letter_count = letters.count(letter)
        dictionary[letter] = letter_count
    print(dictionary)
counting_dicts('programming')
```
