```python
# Assignment 1
class Countdown:
    def __init__(self, number):
        self.number = number
    def __iter__(self):
        return self
    def __next__(self):
        if self.number>=0:
            current_value=self.number
            self.number-=1
            return current_value
        else:
            raise StopIteration
        
for num in Countdown(-5):
    print(num)
```

```python
# Assignment 2
class MyRange:
    def __init__(self, start, stop, step):
        self.start = start
        self.stop = stop
        self.step = step
    def __iter__(self):
        return self
    def __next__(self):
        if self.start<self.stop:
                current_number =self.start
                self.start+=self.step
                return current_number
        elif self.start>self.stop:
                current_number =self.start
                self.start+=self.step
                return current_number
        else:
            raise StopIteration
   

for num in MyRange(10,0,-2):
    print(num)
```

```python
# Assignment 3
def fibonacci(limit):
    a,b = 0, 1
    for _ in range(limit):
        yield a
        a,b = b, a+b

        

for num in fibonacci(10):
    print(num)
```

```python
# Assignment 5
def even_numbers(limit):
    for num in range(0,limit+1):
        if num%2==0:
            yield num

even = even_numbers(20)

def squares(even):
    for even_num in even:
        yield even_num**2

for val in squares(even_numbers(20)):
    print(val)

```

```python
# Assignment 6
from datetime import datetime
import functools
def time_it(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        start = datetime.now()
        result = func(*args, **kwargs)
        stop = datetime.now()
        time_taken = stop - start
        print(f"Function {func.__name__} took {time_taken} to execute")
        return result
    return wrapper

@time_it
def factorial(limit):
    def factorial_recursive(current_point):
        if current_point == 1 or current_point==0:
            return 1
        else:
            return current_point * factorial_recursive(current_point-1)    
    return factorial_recursive(limit)
print(factorial(5))
```

```python
# Assignment 8
def uppercase(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result.upper()
    return wrapper

def exclaim(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result+"!"
    return wrapper

@uppercase
@exclaim
def greetings():
    return "Hello, Hi"

print(greetings())
```

```python
# Assignment 9
def singleton(cls):
    instances ={}
    def wrapper(*args, **kwargs):
        if cls not in instances:
            instances[cls] = cls(*args, **kwargs)
        return instances[cls]
        
    return wrapper

@singleton   
class DatabaseConnection:
    def __init__(self):
        print("Database connection Successful")

db1 = DatabaseConnection()
db2 = DatabaseConnection()

print(db1 is db2)
```

```python
# Assignment 10
def uppercase(cls):
    def wrapper(string):
        return cls(string.upper())
    return wrapper

@uppercase
class ReverseString:
    def __init__(self, string):
        self.string = string
        self.index = len(string)
    def __iter__(self):
        return self
    def __next__(self):
        self.index-=1
        if self.index < 0:
            raise StopIteration
        else:
            
            return self.string[self.index]

operation = ReverseString("hello hi")
for letter in operation:
    print(letter)
```

```python
# Assignment 11
def counter(start):
    while True:
        yield start
        start+=1

func = counter(1)
for _ in range(10):
    print(next(func))
```

```python
# Assignment 12
def safe_divide(lst, divisor):
    lst = iter(lst)
    for num in lst:
        try:
            yield num/divisor
        except ZeroDivisionError:
            yield "You cannot divide by a zero"
            return

lst = [1,2,3,4,5,6]
func = safe_divide(lst, 2)
print(next(func))
print(next(func))
for i in func:
    print(i)
```

```python
# Assignment 13
def open_file(file_name, mode):
    def decorator(func):
        def wrapper(*args, **kwargs):
            with open(file_name, mode) as f:
                return func(f, *args, **kwargs)
        return wrapper
    return decorator

@open_file("test.txt", 'w')
def write_to_file(file, content):
    file.write(content)

write_to_file("Hello, this file was managed by a decorator!")
```

```python
# Assignment 14
class InfiniteCounter:
    def __init__(self, start):
        self.start = start
    def __iter__(self):
        return self
    def __next__(self):
        self.start+=1
        return self.start

result = InfiniteCounter(0)
for _ in range(10):
    print(next(result))
```

```python
# Assignment 15
def integers():
    for i in range(1,11):
        yield i

def doubles(integers):
    for n in integers:
        yield n*2

def negatives(doubles):
    for n in doubles:
        yield -n

int_gen = integers()
double_gen = doubles(int_gen)
final_gen = negatives(double_gen)

for value in final_gen:
    print(value)
```
