##### Map() Function
Applies a functions to all elements in an iterable


Lambda Function With Map

```python
numbers = [1,2,3,4,5,6,7,8]
list(map(lambda x:x**2, numbers))
```

```python
# Mapping Multiple Iterables
numbers1 = [1,2,3]
numbers2 = [4,5,6]
list(map(lambda x,y : x+y, numbers1, numbers2))
```

```python
# Applying map function to a dictionary
def get_name(person):
    return person['name']
people = [
    {'name' : 'Krish', 'age' : 32},
    {'name' : 'Jack', 'age' : 33}
]

list(map(get_name, people))
```
