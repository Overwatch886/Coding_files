```python
## Positional arguments
def print_numbers(*args):
    for number in args:
        print(number)
print_numbers(1,1,2,4,5,3,'gray','player')
```

### Keyword arguments is given by **
In keyword arguments, all arguments will be in the form of key value pairs


```python
# Keywords Arguments
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key} : {value}")
```

```python
print_details(hello = "kid", age = "5")
```

Using Positional and Keyword Arguments Together

```python
def print_details(*args, **kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")
    for number in args:
        print(number)
```

Positional arguments must always come before keyword arguments both when creating the function and calling the function.

```python
print_details(234,1,2,3,4,4,5,45,hello = "kid", age = "5")
```

### example 1 : temperature conversion

```python
def convert_temperature(temp, unit):
    """This function convert temperature between Celcius and Fahrenheit"""
    if unit=='c':
        # Converting celcius to fahrenheit
        return (temp/5*9) + 32
    elif unit=='f':
        # Converting fahrenheit to celcius
        return ((temp - 32)/9)*5
    else:
        print('Invalid Unit')
    
print(convert_temperature(100, 'f'))
print(convert_temperature(37, 'c'))
```

### Password Strength Checker

```python
def is_strong_password(password):
    """This function checks if the password is wrong"""
    # checks if the password has more than 8 characters
    if len(password) < 8:
        return False
    # checks for the presence of a digit
    if not any(char.isdigit() for char in password):
        return False
    # Checks for the presence of lower case
    if not any(char.islower() for char in password):
        return False
    # Checks for the presence of upper case
    if not any(char.isupper() for char in password):
        return False
    # Checks for the presence of special characters
    if not any(char in '!@#$%^&*()_+-' for char in password):
        return False
    return True
   
    
```

```python
is_strong_password('Passw2wuuw$d')
```

Calculate the factorial of a number using recursion

```python
def factorial(n):
    if n==0:
        return 1
    else :
        return n*factorial(n-1)
    
factorial(5)
```
