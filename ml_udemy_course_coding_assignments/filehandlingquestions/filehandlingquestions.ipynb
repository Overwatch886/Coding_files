```python
#Assignment 1
def read_file(filename):
    with open(filename, 'r') as file:
        for lines in file:
            print(lines.strip())
read_file('sample.txt')
```

```python
# Assignment 2
def write_file(strings):
    with open('output.txt', 'w') as file:
        for line in strings:
            file.write(line+'\n')
strings = ['i am', 'who', 'i am']
write_file(strings)
```

```python
# Assignment 3
def copy_file(source, destination):
    with open(source, 'r') as file:
        content = file.readlines()
        print(content)
    with open(destination, 'w') as file:
        file.writelines(content)
source = 'source.txt'
destination = 'destination.txt'
copy_file(source, destination)
```

```python
# Assignment 4
def append_to_file(string):
    with open('log.txt', 'a') as file:
        file.write(string+ '\n')
string = 'I am appending a given string'
append_to_file(string)
```

```python
# Assignemnt 5
def doc_word_counter(doc):
    with open(doc, "r") as file:
        content = file.read()
        content = content.split()
        print(content)
        word_count = len(content)
        return word_count
doc_word_counter("document.txt")
```

```python
# Assignment 6
def word_replacer(doc, word_to_replace, replacement):
    with open(doc, "r+") as file:
        content = file.read()
        print(content)
        new_content = content.replace(word_to_replace, replacement)
        print(new_content)
        file.seek(0)
        file.write(new_content)
word_replacer("document.txt", "will", "would")
```

```python
# Assignment 7
def read_in_reverse(doc):
    with open(doc, 'r', newline='') as file:
        content = file.readlines()
        print(content)
        for line in content[::-1]:
            print(line.strip())
read_in_reverse("document.txt")
```

```python
# Assignment 8
# Version 1
# Excluding white space characters(tabs, nelines and spaces)
def line_word_char_counter(doc):
    total_char_count=0
    total_word_count=0
    no_of_words=0
    line_no=0
    with open(doc, "r", newline="") as file:
        lines = file.readlines()
        no_of_lines = len(lines)
        print(f"There are {no_of_lines} lines in the doc")
        for line in lines:
            total_word_count+= no_of_words
            
            words= line.split()
            no_of_words = len(words)
            line_no+=1
            print(f"There are {no_of_words} words in line {line_no}")
            print(words)
            word_no=0
            for word in words:
                word_no+=1
                char_count=0
                for letter in word:
                    print(letter)
                    char_count+=1
                total_char_count+=char_count
                print(total_char_count)
                print(f"There are {char_count} characters in word {word_no} on line {line_no}")
        print(f'''In total, there are 
              {no_of_lines} lines
              {total_word_count} words and
               {total_char_count} characters.
                 ''')
line_word_char_counter("destination.txt")
```

```python
# Assignment 8
# Version 2
# Excluding white space characters(tabs, nelines and spaces)
def line_word_char_counter(doc):
    with open(doc, "r", newline=None) as file:
        lines = file.readlines()
        print(lines)
        words = sum(len(line.split()) for line in lines)
        chars = sum(len(ch) for line in lines for ch in line if not ch.isspace())
    return len(lines), words, chars
line_word_char_counter("destination.txt")
```

```python
# Assignment 8
# Version 3
# Including white space characters(tabs, nelines and spaces)
def line_word_char_counter(doc):
    with open(doc, "r", newline=None) as file:
        lines = file.readlines(10)
        print(lines)
        words = sum(len(line.split()) for line in lines)
        chars = sum(len(line) for line in lines)
    return len(lines), words, chars
line_word_char_counter("destination.txt")
```

```python
# Assignment 9
def file_merger(doc1, doc2):
    with open(doc1, 'r') as file1:
        with open(doc2, 'r') as file2:
            file1_content = file1.read()
            file2_content = file2.read()
            merged_content = file1_content + file2_content
    with open("merged.txt", 'w') as file:
        file.write(merged_content)
file_merger('document.txt', 'destination.txt')
```

```python
# Assignment 10
def file_splitter(doc_to_split):
    with open(doc_to_split, 'r') as file:
        lines_to_read=100
        contents = file.readlines()
        no_of_lines = len(contents)   
        doc_no=1
        for line in range(0, no_of_lines, lines_to_read):
            new_doc = f"doc{doc_no}.txt"
            with open (new_doc, 'a') as new_file:
                new_file.writelines(contents[line:line+lines_to_read])
            doc_no+=1
                
            

file_splitter('random_text.txt')

```

```python
# Assignment 12
# Binary File Operations
def bin_operations(bin_file):
    with open(bin_file, "rb") as file:
        content= file.read()
    with open("copy_image.bin", "wb") as copy_file:
        copy_file.write(content)
bin_operations("image.bin")
```

```python
# Assignment 13
# CSV File Operations
import csv
def csv_operations(file):
    with open(file, "r") as file:
        content = csv.DictReader(file)
        return list(content)

csv_operations("data.csv")
```

```python
# Assignment 14
# JSON File Outputs
import json
def json_operation(file):
    with open(file, "r") as json_file:
        data = json.load(json_file)
        return data

print(json_operation("data.json"))
```

```python
# Assignment 15
# File Permission Handling
def read_protected_file(file):
    try :
        with open(file, "r") as file:
            content = file.read()
            return content
    except PermissionError as e:
        return f"[ERROR] {e} error occurred"

read_protected_file("destination.txt")
```
