**Question 19:** Write a Python program to check if a string is a palindrome.

```python
# Your code here
_string = "paP"
_string = _string.lower()
reversed_string = ""
for char in _string:
  reversed_string = char + reversed_string
print(reversed_string)
if _string == reversed_string:
  print(f"{_string} is a Palindrome")
else :
  print(f"{_string} is not a Palindrome")
```

**Question 20:** Write a Python program to sort a list of numbers in ascending order.

```python

# Your code here
numbers = [2, 6, 79, 99, 73, 24, 57, 100]
sorted_list = []
tests = 0
original_length = len(numbers)
while tests < original_length:
  min = numbers[0]
  for i in numbers:
    if i <= min:
      min = i
    else :
      continue
  tests = tests + 1
  sorted_list.append(min)
  numbers.remove(min)
  #print(sorted_list)
print(sorted_list)
```
