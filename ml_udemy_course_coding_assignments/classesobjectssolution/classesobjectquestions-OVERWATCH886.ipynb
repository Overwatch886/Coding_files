```python
# Assignment 11
from abc import ABC, abstractmethod
import math
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi* (self.radius**2)
    
class Square(Shape):
    def __init__(self, length):
        self.length = length

    def area(self):
        return (self.length**2)


square = Square(4)
print(square.area())

circle = Circle(7)
print(circle.area())
```

```python
# Assignment 12
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        self.other = other
        return Vector(self.x+other.x, self.y+other.y)
    
    def __str__(self):
        return (f"{self.x}, {self.y}")
    
vector = Vector(2,3)
v2 = Vector(2,3)
print(vector)
v3 =vector+v2

print(v3+vector)
        
```

```python
# Assignment 13
class InsufficientBalanceError(Exception):
    pass

class BankAccount:
    def __init__(self, account_number, balance):
        self.__account_number = account_number
        self.__balance = balance
    def deposit(self, amount):
        self.__balance += amount
        print(f"{amount} deposited")
        print(f"Your new balance is {self.__balance}")
    def withdraw(self, amount):
            try:
                if self.__balance >= amount:
                    self.__balance -= amount
                    print(f"{amount} withdrawn")
                    print(f"Your new balance is {self.__balance}")
                else:
                    raise(InsufficientBalanceError("Insufficient Balance"))
            except InsufficientBalanceError as e:
                print(f"[ERROR] {e}")
    def display_balance(self):
        print(f"Your balance is {self.__balance}")

israel = BankAccount(8056668785, 4000)

israel.withdraw(70000)
```

```python
# Assignment 14
class Filemanager:
    def __init__(self, file, mode):
        self.file = file
        self.mode = mode

    def __enter__(self):
        self.file= open(self.file, self.mode)
        return self.file
            
    def __exit__(self, exc_type, exc_value, traceback):
        self.file.close

with Filemanager('book.txt', "r") as file:
    content = file.read()
    print(content)
```

```python
# Assignment 15
class Calculator:
    def __init__(self, value = 0):
        self.value = value

    def add(self, num2):
            self.value +=num2
            return self

    def subtract(self, num2):
            self.value -= num2
            return self
    
    def multiply(self, num2):
            self.value *= num2
            return self
    
    def divide(self, num2):
            self.value /= num2
            return self
    
num1 = Calculator(10)
ans =num1.add(27).multiply(69).divide(4).add(22)
print(ans.value)
```
