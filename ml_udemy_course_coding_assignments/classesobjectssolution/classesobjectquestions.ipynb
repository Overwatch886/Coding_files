```python
# Assignment 4
class BankAccount:
    def __init__(self):
        self.account_number = 8056668785
        self.balance = 20000
    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} deposited")
        print(f"Your new balance is {self.balance}")
    def withdraw(self, amount):
        self.balance -= amount
        print(f"{amount} withdrawn")
        print(f"Your new balance is {self.balance}")

israel = BankAccount
israel.deposit(200, 5000)
```
