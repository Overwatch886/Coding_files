```python
# Assignment 1 
# Handling Division by Zero
def divide(number1, number2):
    try:
        print(number1/number2)
    except ZeroDivisionError as e:
        print(f"[ERROR] The error {e} occurred")
    finally:
        print("Operation Complete")

number1 = int(input("Enter the first number "))
number2 = int(input("Enter the second number "))
divide(number1, number2)
```

```python
# Assignment 2
def read_file(filename):
    try:
        file = open(filename, "r")
        content = file.read()
        file.close()
        return content
    except FileNotFoundError as e:
        print(f"[ERROR] {e}")
    finally:
        print("File read operation complete")
read_file("data.txt")
```

```python
# Assignment 3
def list_adder(lst):
    try:
        total = sum(lst)
        return total
    except Exception as e:
        print(f"[ERROR] : {e}")
    finally:
        print("Execution Complete")
list_adder([1,2,3,4,5, "Name"])
```

```python
# Assignment 7
def list_elem_finder(lst, index):
    try:
        value = lst[index]
        result = f"The element at index {index} is {value}"
    except IndexError:
        result = "The given index is out of range for this list"
    finally:
        print("Execution Complete")
        return result
list_elem_finder([2,4,56,67,True], 5)
```

```python
# Assignment 8
def visit_url(url):
    import requests
    try:
        value = requests.get(url)
    except requests.RequestException as e:
        value= f"The Error {e} occurred"
    finally:
        print("Execution Complete")
        return value
visit_url("https://www.gle.bag")

```

```python
# Assignment 9
import json
def json_parser(json_str):
    try:
        json_in_dict = json.loads(json_str)
    except json.JSONDecodeError as e:
        json_in_dict= f"Unable to Decode Json. The error {e} occurred"
    finally:
        print("Execution Complete")
        return json_in_dict
json_parser('{"Name": "Israel"}')
```

```python
# Assignment 10
class NegativeNumberError(Exception):
    pass
def checking_for_negatives(lst):
    try:
        for num in lst:
            if num <0:
                raise NegativeNumberError (f"Error : Negative number {num} found")
            else:
                print(num)
    except NegativeNumberError as e:
        print(e)
    finally:
        print("Execution Complete")
checking_for_negatives([1,2,3,4,-5,6,7,-8])
```

```python
# Assignment 11
def risky_function():
    raise ValueError("THis is a risky function")
def safe_function():
    try:
        user_input = input("ENter risky or safe")
        if user_input == "risky":
            risky_function()
        else:
            return "We are safe"
    except ValueError as e:
        return f"Error {e} occured"
    finally:
        print("Execution Complete")
safe_function()
```

```python
# Assignment 13
def str_to_int(lst):
    for str in lst:
        try:
            result = int(str)
        except ValueError:
            result = f"String {str} could not be converted to integer"
        finally:
            print(result)
    
str_to_int(['str', 'sf', 'safe', 'favour', 'stay'])
```

```python
# Assignment 15
def write_str_to_file(lst, file):
    try:
        with open(file, 'w') as _file:
            for string in lst:
                _file.write(string+'\n')
    except IOError:
        print("IOERROR")
    finally:
        print("Execution Complete")
lst = ['str\n', 'sf\n', 'safe\n', 'favour\n', 'stay\n']
write_str_to_file(lst, "data.txt")
```
