```python
# Assignment 12
num = int(input("Enter the factorial you want to calculate :"))
factorial = num
if num != 0:
  while num > 1:
    factorial = factorial * (num-1)
    num = num-1
  print(factorial)
else :
  print("1")
```

```python
# Assignment 13
num = int(input("Enter the sum you want to calculate :"))
factorial = num
if num != 0:
  while num > 1:
    factorial = factorial + (num-1)
    num = num-1
  print(factorial)
else :
  print("0")
```

```python
# Assignment 14
num = int(input("Input Number"))
if num%2 != 0:
  for i in range(2, num):
    if i%2 == 0:
      continue
    elif num%i == 0:
      print("This is not prime")
    else :
      print("This is prime")
      break
else :
  print("This is not prime")
```
