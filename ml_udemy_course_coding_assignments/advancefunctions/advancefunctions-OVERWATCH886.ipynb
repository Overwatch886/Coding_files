```python
# Assignment 1

def fibonacci(n, memo={}):
    if n<= 1:
        return n
    if n in memo:
        return memo[n]
    
    memo[n] = fibonacci(n-1, memo) + fibonacci(n-2, memo)
    return memo[n]
    
    
print(fibonacci(10))
```

```python
# Assignment 6
from datetime import datetime
def funcdurationchecker(func):
    def wrapper(*args, **kwargs):
            starttime = datetime.now()
            result = func(*args, **kwargs)
            endtime = datetime.now()
            print(f'The function {func.__name__} took {endtime-starttime} time to execute')
            return result
    return wrapper




    
```

```python
# Assignment 6
@funcdurationchecker
def fibonacci(n, memo={}):
    return sum(x**2 for x in range(n))
    
print(fibonacci(10))
#
# print(fibonacci(20))
```

```python
# Assignmment 7
def higherorderfunc(filter_func, map_func, lst):

    filtered = list(filter(filter_func, lst))
    mapped = list(map(map_func, filtered))
    return mapped

higherorderfunc(lambda num:num%2==0, lambda x:x**2, [1,2,3,4,5])
```

```python
# Assignment 13
def context_manager(lst):
    try:
        file = open("output.txt", 'w')
        for num in lst:
            num = str(num)
            file.write(num+'\n')
        file.close()
    except IOError as e:
        print(f"An error {e} occurred")

context_manager([1,2,3,4,5,5,6,7,7])
```

```python
# Assignment 13
def context_manager(lst):
    try:
        lst = [str(num)+'\n' for num in lst]
        with open("output.txt", 'w') as file:
            file.writelines(lst)
    except IOError as e:
        print(f"An error {e} occurred") 

context_manager([1,2,3,4,5,6,7,8])
```

```python
# Assignment 14
strings =[]
booleans = []
integers =[]
def sorter(lst):
    for obj in lst:
        if type(obj) == int:
            integers.append(obj)
        elif type(obj) == bool:
            booleans.append(obj)
        elif type(obj) == str:
            strings.append(obj)
    return integers, booleans, strings

sorter([1,2,3,4,4,5,6,7,8, True, False, True, "play", "hide"])
```
