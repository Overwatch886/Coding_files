```python
# Assignment 1
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

class Dog(Animal):
    def __init__ (self, name, species, breed):
        super().__init__(name, species)
        self.breed = breed
dog = Dog("Clowey", "Chiwawa", "wolverine")
print(dog.species)
```

```python
# Assignemnt 3
class Dog(Animal):
    def __init__ (self, name, species, breed):
        super().__init__(name, species)
        self.breed = breed
    def __str__(self):
        return f"You have a dog named {self.name} of {self.species} specie and breed {self.breed}"
    def __init__ (self, name, species, breed):
        super().__init__(name, species)
        self.breed = breed
    def __str__(self):
        return f"You have a dog named {self.name} of {self.species} specie and breed {self.breed}"
    def bark(self):
        return "Woof"
    
dog = Dog("Clowey", "Chiwawa", "wolverine")
dog.bark()
```

```python
# Assignment 4
class Walker:
    def walk(self):
        print(f"I am walking")

class Runner:
    def run(self):
        print(f"I am running")
    
class Athlete(Walker, Runner):
    pass

athlete = Athlete()
athlete.walk()
athlete.run()

```

```python
# Assignment 5
class Athlete(Runner, Walker):
    def walk(self):
        print(f"I can still walk")
        super().walk()

athlete2 = Athlete()
athlete2.walk()
```

```python
# Assignment 6
class Athlete(Runner, Walker):
    def __init__(self, training_hours):
        self.training_hours = training_hours
        super().__init__()
    def train(self):
        print(f"You have trained for {self.training_hours} hours")

athlete3 = Athlete(30)
athlete3.train()
```

```python
# Assignment 7
class A:
    def show(self):
        print("I am show")

class B(A):
    def show(self):
        print("I am the new show, B")

class C(A):
    def show(self):
        print("I am the new show, C")

class D(B, C):
    pass

d = D()
d.show()
```

```python
# Assignment 8
class Shape:
    def __init__(self, color):
        self.color = color

class Circle(Shape):
    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius

circle = Circle("red", 34)
print(circle.color)
print(circle.radius)
```

```python
# Assignment 9
class Person:
    def __init__(self, name):
        self.name = name
class Employee:
    def __init__(self, employee_id):
        self.employee_id = employee_id

class Manager(Person, Employee):
    def __init__(self, name, employee_id):
        super().__init__(name)
        Employee.__init__(self, employee_id)

manager = Manager("Israel", 886)

print(manager.name, manager.employee_id)
```

```python
# Assignment 10
class Vehicle:
    def start(self):
        print(f"The vehicle starts")
    
class Car(Vehicle):
    def start(self):
        print(f"the car starts")
        super().start()

car = Car()
car.start()
```

```python
# Assignment 11
class Flyer:
    def fly(self):
        print(f"A flyer flys")

class Swimmer:
    def swim(self):
        print("A swimmer swims")
    
class Superhero(Flyer, Swimmer):
    pass
superhero = Superhero()

superhero.fly()
superhero.swim()
```

```python
# Assignment 12
class Base1:
    def __init__(self, a):
        self.a = a

class Base2:
    def __init__(self, b):
        self.b = b

class Derived(Base1, Base2):
    def __init__(self, a, b, c):
        super().__init__(a)
        Base2.__init__(self, b)
        self.c = c
        

derived =  Derived(1,2,3)
print(derived.a, derived.b, derived.c)
```

```python
# Assignment 13
class Animal:
    pass

class Cat(Animal):
    pass

animal = Animal()
cat = Cat()
print(isinstance(animal, Cat))
print(isinstance(animal, Animal))
print(isinstance(cat, Cat))
print(isinstance(cat, Animal))
```

```python
# Assignment 14
class Bird:
    def speak(self):
        print("A Bird speaks")

class Parrot(Bird):
    def speak(self):
        print("A Parrot speaks")

class Penguin(Bird):
    def speak(self):
        print("A Penguin speaks")

parrot = Parrot()
penguin = Penguin()

birds = [parrot, penguin]
for bird in birds:
    bird.speak()
```

```python
# Assignment 15
class Device:
    def __init__(self, brand):
        self.brand = brand

class Phone(Device):
    def __init__(self, brand, model):
        self.model = model
        super().__init__(brand)

class Camera:
    def __init__(self, resolution):
        self.resolution = resolution

class Smartphone(Phone, Camera):
    def __init__(self, brand, model, resolution):
        super().__init__(brand, model)
        Camera.__init__(self, resolution)

smartphone = Smartphone("Nokia", "T10", "15MP")
print(smartphone.brand, smartphone.model, smartphone.resolution)
```
