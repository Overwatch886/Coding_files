```python
password = input("Enter your password")
special_chars = '!@#$%^&*()'
numbers = '0123456789'
strength = 'weak'
for char in special_chars:
    if char in password:
        for number in numbers:
            if number in password:
                    strength = 'strong'
                    break
if strength == 'strong':
     print("Strong Password") 
else:
     print("Weak Password")   
```

```python
# Hangman
trials_left = 8
print("You have 8 trials in this game")
word = ['i', 's', 'r', 'a', 'e', 'l', 'a'] 
word_count = len(word)
while trials_left!= 0 and word_count!=0:
    print(f"You have {word_count} letters to guess")
    player_guess = input("Guess a letter ")
    if player_guess in word:
        while player_guess in word:
            word.remove(player_guess)
        print(f"You guessed the letter {player_guess} correctly")
        word_count = len(word)
        
     
    else:
        trials_left-=1
        print(f"Wrong, you have {trials_left} trials left")
if word_count == 0:
    print("You win")
else:
    print("You Lose")
    
```

```python
# Implementing s contsct book
contacts = {'09165126149': "Israel", '08161428643': "Dayo"}

def add_contacts(name, number):
    # This is used for adding and editing contact as well
    contacts[name] = number
def delete_contacts(number):
    del contacts[number]


print("What do you want to do")
user_intent = (input("Do you want to add, delete or review contacts")).lower()
if user_intent == "add":
    name = input("Enter the name of the contact")
    number = input("Enter the mobile number of the contact")
    add_contacts(name, number)
    print("Here ia your updated contacts list")
    print(contacts)
elif user_intent == "delete":
    number = input("Enter the mobile number of the contact")
    delete_contacts(number)
    print("Here ia your updated contacts list")
    print(contacts)
elif user_intent == "review":
    print(contacts)
else:
    print("Invalid operation")
```
