```python
# 1 Personal Greetings
username = input("Enter your name")
print(f"Hello {username}! Welcome to our store")
```

```python
# 2 Swapping Variables
unit1 = input("Entrer unit1")
unit2 = input("Enter unit2")
var_holder=unit1
unit1 = unit2
unit2=var_holder
print(f"{unit2} : unit2")
print(f"{unit1} : unit1")
```

```python
# 3  Shopping cart total
price_holder = 0
for price in range(1,4):
    prices = int(input(f"Enter the cost of item {price}"))
    price_holder += prices 

print(price_holder)
```

```python
# 4 Discount Calculator
original_price = int(input("Enter Original Price "))
discount_percentage = int(input("Enter discount percentage "))
discounted_price = ((100 - discount_percentage) * original_price)/100
print(discounted_price)
```

```python
# 5 Temperature Conversion
celsius_temp = int(input("Enter the temperature in celsius "))
fah_temp = (celsius_temp*9/5) + 32
print(f"{celsius_temp}℃ in fahrenheit is {fah_temp}℉")
```

```python
# 6 Currency Converter
currency_in_dollars = float(input("Enter the currency value in dollars. It would be converted to naira based on current interest rates"))
rates = 1457.08
currency_in_naira = currency_in_dollars * rates
print(f"At the current rate of #{rates} per dollar, ${currency_in_dollars} is worth #{currency_in_naira} ")
```

```python
# 7 Average Speed Calculator
distance = float(input("Enter the distance in kilometers "))
time = float(input("Enter the time in hours "))
average_speed = distance/time
print(f"Average speed is {average_speed}km/hr")
```
