##### Assignment 2

Map function is use for applying  a function to an iterable.

```python
# Using map function and lambda to convert a list of temperatures in celcius to fahrenheit
celsius_top_fahrenheit_converter = lambda temp : 9 * (temp/5) + 32
temperatures = [0,10,20,30,40]
result = map(celsius_top_fahrenheit_converter, temperatures)
print(list(result))
```

##### Assignment 3

Using the sorted function

```python
# Using sorted with a lambda function to sort a list by last character
lst = ['dog', 'cat', 'lion', 'tiger']
sorted_list = sorted(lst, key = lambda x : x[-1], reverse = False)
print(sorted_list)      
```

Using the filter function

```python
# Using filter with a lambda to get all words longer than 4 letters
lst = ['data', 'science', 'AI', 'python']
result = filter(lambda word: len(word) > 4 , lst)
print(list(result))
```
