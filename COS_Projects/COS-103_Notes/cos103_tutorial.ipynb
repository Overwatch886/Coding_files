```python
## Creating a mini calculator
def mini_calc(num1,num2):
    user_operation = input('Enter the opertion to be performed')
    if user_operation.lower() == 'add':
        return num1 + num2
    elif user_operation.lower() == 'multiply':
        return num1 * num2
    elif user_operation.lower() == 'subtract':
        return num1 - num2
    elif user_operation.lower() == 'divide':
        return num1 / num2
    else :
        return ("Invalid Operation")
result = mini_calc(8,6)
print(result)
```

```python
def factorial(num): 
    if num==0:
        return 1
    else :
        tracker = num-1
        while tracker!=0:
            num = num*tracker
            tracker-=1
        return num
factorial(0)
```

With a class in python, we can create different identites of that particular class

 To create a class in python

```python
class Dog:
    ## Initializes the class and give it its properties using __init__
    def __init__(self,name,fav_food):
        self.name = name
        self.favourite_food = fav_food
        age = 24

    def __str__(self):
        return(f"{self.name} likes {self.favourite_food}")
    
    def bark(self):
        return "Woff Woff"
    
    def jump(self,num):
        while num>0:
            print("Jump")
            num-=1
        return f"{self.name} has finished jumping after eating {self.favourite_food}"
tolu_dog = Dog("Bingo", "pasta")
print(tolu_dog.name)
print(tolu_dog.favourite_food)
print(tolu_dog.bark())
kissi_dog = Dog("Jake", "Souls")
kissi_dog.jump(7)
print(kissi_dog)
```

Assignment : Write a python program that defines a class bank account, that has properties of an owner, and a balance, deposit method, withdraw method, get balance method, a __Str__ method 

```python
class bank_account:
    def __init__(self, owner):
        self.owner = owner
        self.balance = 0
    def deposit(self, amount):
        self.balance += amount
        return f"N{amount} has been successfully deposited \nNew balance is {self.balance}"
    def withdraw(self, amount):
        if self.balance > amount :
            self.balance = self.balance - amount
            return f"N{amount} has been successfully withdrawn from available balance \nNew balance is {self.balance}"
        else : 
            return("Insufficient balance")
    def get_balance(self):
        return self.balance
    def __str__(self):
        return(f"{self.owner} Welcome to my bank")

my_account = bank_account("Israel")
print(my_account)
print(my_account.withdraw(1000))
print(my_account.get_balance())
print(my_account.deposit(5000))
print(my_account.withdraw(1000))
```
