```python
## Creating a mini calculator
def mini_calc(num1,num2):
    user_operation = input('Enter the opertion to be performed')
    if user_operation.lower() == 'add':
        return num1 + num2
    elif user_operation.lower() == 'multiply':
        return num1 * num2
    elif user_operation.lower() == 'subtract':
        return num1 - num2
    elif user_operation.lower() == 'divide':
        return num1 / num2
    else :
        return ("Invalid Operation")
result = mini_calc(8,6)
print(result)
```
